```python
# StreamListener class inherits from tweepy.StreamListener and overrides on_status/on_error methods.
class StreamListener(tweepy.StreamListener):
    def clean_tweet(self, tweet):
        '''
        Utility function to clean tweet text by removing links, special characters
        using simple regex statements.
        '''
        return ' '.join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t]) | (\w+:\ / \ / \S+) |(?<!\s[\dA-Z]),(?!\s+\d,?)", " ",
                               tweet).split())

    def get_tweet_sentiment(self, tweet):
        '''
        Utility function to classify sentiment of passed tweet
        using textblob's sentiment method
        '''
        # create TextBlob object of passed tweet text
        analysis = TextBlob(self.clean_tweet(tweet))
        # set sentiment
        if analysis.sentiment.polarity > 0:
            return 1
        elif analysis.sentiment.polarity == 0:
            return 0
        else:
            return -1

    def on_status(self, status):

        print(status.id_str)
        

        # remove characters that might cause problems with csv encoding
        tweet = self.clean_tweet(status.text)
        count = 1
        data = {
            "id": status.id_str,
            "tweet": tweet,
            "label": self.get_tweet_sentiment(tweet)
            
        }
     

        with open("train.csv", "a", encoding='utf-8') as f:
            f.write("%s,%s,%s\n" % (status.id_str, self.get_tweet_sentiment(tweet), tweet))
            count += 1

    def on_error(self, status_code):
        print("Encountered streaming error (", status_code, ")")
        sys.exit()
```

```python
if __name__ == "__main__":
    

    # initialize stream
    streamListener = StreamListener()
    stream = tweepy.Stream(auth=api.auth, listener=streamListener)

    with open("train.csv", "w", encoding='utf-8') as f:
        f.write("id,label,tweet\n")
    alert_flood = [' flood ', ' flooding ', ' flooded ', ' surge ', ' torrent ']
    alert_storm = [' storm ', ' tornado ', ' wind ', ' snow ', ' hale ', ' rain ', ' deluge ', ' hurricane ', ' avalanche ', ' ice ']
    alert_shooting = [' shooting ', ' shooter ', ' shots fired ', ' bullet ', ' gunshot ', ' murder ', ' driveby ']
    alert_fire=['fire']
    alert_accident = ['accident']
    alert_any = alert_fire + alert_flood + alert_storm + alert_shooting + alert_accident + alert_fire
    stream.filter(track=alert_any, languages=language)
```
